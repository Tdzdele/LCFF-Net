```python
import pandas as pd
import matplotlib.pyplot as plt
filenames = ['', '', '', '', '']
def plot_csv_comparison(filenames):
    colors = ["#000000","#ff0000","#00ff00","#0000ff","#ff00ff","#808000",]
    for filename, color in zip(filenames, colors):
        if filename:
            df = pd.read_csv(f'{filename}/results.csv')
            x, y = df.iloc[:, 0], df.iloc[:, 7] * 100
            plt.plot(x, y, label=filename, color=color)
    plt.xlabel('Epochs')
    plt.ylabel('mAP50')
    plt.title('Comparison of Model Training')
    plt.legend()
    plt.grid(True)
    plt.show()

filenames = [
            'runs/train/LCFF-Net/LCFF-Net-t',
            'runs/train/Attention/LCFF-Net-t-EMA',
            'runs/train/Attention/LCFF-Net-t-CAA',
            'runs/train/Attention/LCFF-Net-t-MLCA',
            'runs/train/Attention/LCFF-Net-t-AA',
            ]
plot_csv_comparison(filenames)
```
